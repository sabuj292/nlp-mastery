```python
docs = [
    "I love NLP",
    "NLP loves Python",
    "I love Python"
]
```

### Step 1: Normalize (lowercase + split)

```python
tokenized_docs = [doc.lower().split() for doc in docs]
tokenized_docs
```

#### # Step 2: Count document frequency

```python
df_counts = {}
for tokens in tokenized_docs:
    unique_tokens = set(tokens)
    for token in unique_tokens:
        df_counts[token] = df_counts.get(token, 0) + 1
        
print("Document Frequency of each word:")
for word, count in df_counts.items():
    print(word, ":", count)
```

Step 3. Visualize Top 10 Words

```python
import matplotlib.pyplot as plt

# Sort words by frequency
sorted_df = sorted(df_counts.items(), key=lambda x: x[1], reverse=True)[:10]

words, counts = zip(*sorted_df)

plt.bar(words, counts)
plt.title("Top 10 Words by Document Frequency")
plt.xlabel("Words")
plt.ylabel("Document Frequency (# of docs)")
plt.show()
```
